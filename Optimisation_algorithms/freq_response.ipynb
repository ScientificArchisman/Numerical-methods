```python
import numpy as np 
import matplotlib.pyplot as plt
plt.style.use(["science", "notebook", "grid", "dark_background"])
```

## LCR Circuit

```python
Vin, Vout, freq = np.loadtxt("data/lcr.txt", skiprows=1, unpack=True)
popt = np.polyfit(x = freq, y = Vout/Vin, deg=9)
freq_fit = np.linspace(2.5, 13, 1000)
fit_data = np.polyval(p = popt, x = freq_fit)
resonant_freq = freq_fit[np.argmax(fit_data)]
```

```python
plt.figure(figsize = (15, 8))
plt.plot(freq_fit, fit_data, "-.", label = "Fitting Curve")
plt.scatter(freq, Vout/Vin,  edgecolor = "white", s = 50, facecolor = "red", label = "Data points")
plt.axvline(x = resonant_freq, color = "lightgreen", label = f"Resonant Freq = {resonant_freq:.3f} kHz")
plt.locator_params(axis = "both", nbins = 20)
plt.xlabel(r"Frequency(kHz) $\rightarrow$", fontsize = 15, fontweight = "bold")
plt.ylabel(r"$\frac{V_{out}}{V_{in}} \times 10^3 (V) \rightarrow$ ", fontsize = 15, fontweight = "bold")
plt.title("LCR Circuit response", fontsize = 20, fontweight = "bold")
plt.legend()
```

```python
L = 33 * 10**(-3)
R = 1000
tau = L / R
cutoff_freq = 1 / tau
```
